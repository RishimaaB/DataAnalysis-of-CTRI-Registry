```python
#Program 1
#For extracting zipcodes from address
# This is Final
import pandas as pd
import re

# Load the CSV file into a DataFrame
input_xlsx = 'input_163list_zipcodes.xlsx'
df = pd.read_excel(input_xlsx)

# Function to extract 6-digit zip codes from an address
def extract_zip_code(address):
    # Updated regex pattern to match 6-digit zip codes with or without spaces
    #zip_code_pattern = r'\b\d{6}\b|\b\d{3}\s?\d{3}\b'
    zip_code_pattern = r'\d{6}|\d{3}\s?\d{3}'
    zip_codes = re.findall(zip_code_pattern, address)
    return ', '.join(zip_codes) if zip_codes else ''

# Apply the function to the 'PI_Address_cleaned' column and create a new 'ZipCode' column
df['ZipCode'] = df['PI_Address'].apply(extract_zip_code)

# Save the DataFrame with extracted zip codes to a new CSV file
output_xlsx = 'extracted_zip_codes_163list.xlsx'
df.to_excel(output_xlsx, index=False)

print(f"6-digit zip codes have been extracted and saved to '{output_xlsx}'.")
```

```python
#Program 2
#Removing all the special characters and saving it to a new file.
import pandas as pd
import re

# Read the Excel file into a DataFrame
df = pd.read_excel('RB_163list_affiliation.xlsx')  # Replace 'your_file.xlsx' with the actual file path

# Define a function to remove special characters using regex
def remove_special_characters(text):
    # Use regex to remove all characters that are not letters, numbers, or whitespace
    return re.sub(r'[^a-zA-Z0-9\s]', '', str(text))

# Apply the function to the column containing text with special characters
df['PI_Affiliation_cleaned'] = df['PI_Affiliation_cleaned'].apply(remove_special_characters)

# Save the cleaned DataFrame back to the Excel file
df.to_excel('RB_Affiliation_spclchars_rmvd_163_25jan2024.xlsx', index=False)  # Replace 'cleaned_file.xlsx' with your desired output file path

print("Removed all, Thank you")
```

```python
#Program 3
#For arranging names in the order of commas subsequently
import openpyxl

# Specify the input Excel file name
input_file_name = 'input_122_1136vs2959.xlsx'

# Specify the output Excel file name
output_file_name = 'output_122_1136vs2959.xlsx'

# Open the Excel file and read the data
workbook = openpyxl.load_workbook(input_file_name)
sheet = workbook.active

# Extract data from the Excel sheet
data = [name.strip() for row in sheet.iter_rows(min_row=1, values_only=True) for name in row]
output_data = [name for names in data for name in names.split(',')]

# Create a new workbook for the output
output_workbook = openpyxl.Workbook()
output_sheet = output_workbook.active

# Write the output data to the new workbook
for index, name in enumerate(output_data, start=1):
    output_sheet.cell(row=index, column=1, value=name)

# Save the output workbook
output_workbook.save(output_file_name)

print("Done")
```

```python
#Program 4
#This code is used for ordering the records associated with a particular name in another list (This gives the same result as 
#excel's match functions)
import pandas as pd
## Reading two excel files
data1 = pd.read_excel('output_185_69_1136.xlsx',sheet_name= '69')
data2 = pd.read_excel('3706.xlsx',sheet_name= '3706')

data1['PI_Name_No_spaces'] = data1['PI_Name_No_spaces'].str.lower()
data2['PI_Name_No_spaces'] = data2['PI_Name_No_spaces'].str.lower()

# Using merge function by setting how='inner'
output1 = pd.merge(data1, data2, 
                   on='PI_Name_No_spaces', 
                   how='inner')

##To print the result to an output file
output1.to_excel("Merging_the_names_75list_trial2.xlsx", index=False)
  
## Display result
print("Done")
```
